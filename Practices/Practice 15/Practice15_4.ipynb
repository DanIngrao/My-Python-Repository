Creamos dos listas, x e y. Llenamos a la lista x de valores de 1 al 100.
Los valores de y van a equivaler al cuadrado del respectivo valor en x con el mismo indice.

```python
x = list(range(101))
y = []
for numero in x:
    y.append(numero**2)
```

Graficamos ambas listas dentro de un grafico

```python
plt.plot(x,y)
```

Hay otra manera de crear graficos en Matplotlib, utilizando el metodo orientado a objetos (POO)

```python
fig, ax = plt.subplots()

ax.plot(x,y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = list(range(101))
y = []
for numero in x:
    y.append(numero**2)

fig, ax = plt.subplots()

ax.plot(x,y)

ax.set(title='Grafico de casos de covid-19 en Latam',xlabel='Dias',ylabel='Casos confirmados')

fig.savefig("ejemplo-grafico-covid.png")
```

Graficos de dispercion

```python
import numpy as np

x_1 = np.linspace(0,100,20)

y_1 = x_1 ** 2

fig, ax = plt.subplots()
ax.scatter(x_1,y_1)
```

```python
fig, ax = plt.subplots()

x_2 = np.linspace(-10, 10, 100)
y_2 = np.sin(x_2)

ax.scatter(x_2,y_2)
```

Creamos un diccionario con tres platos y su respectivo precio

```python
comidas = {'lasagna': 350,'sopa':150,'roast beef':650}

fig, ax = plt.subplots()

ax.bar(comidas.keys(), comidas.values())

ax.set(title='Precios de comidas',xlabel='Comidas',ylabel='Precios')
```

```python
fig, ax = plt.subplots()
ax.barh(list(comidas.keys()),list(comidas.values()))
ax.set(title='Precios de comidas',xlabel='Precios',ylabel='Comidas')
```

Histograma

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()

ax.hist(x)
```

Creamos una figura con 4 subgraficos (2 por fila)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(20,5))
```

Creamos la misma disposicion de graficos, con un tamanio de figura de 10x5

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Para nuestro primer grafico, tomamos el conjunto x_1, y_1, y generamos un grafico de lineas

ax1.plot(x_1,y_1)

# Para nuestro segundo grafico, tomamos el conjunto x_2, y_2, y generamos un grafico de dispersion

ax2.scatter(x_2,y_2)

# Creamos un grafico de barras

ax3.bar(comidas.keys(),comidas.values())

# Por ultimo un histograma

ax4.hist(np.random.randn(1000))
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

Habiendo cambiado el estilo, lo probamos

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Para nuestro primer grafico, tomamos el conjunto x_1, y_1, y generamos un grafico de lineas

ax1.plot(x_1,y_1, color='#fcba05')

# Para nuestro segundo grafico, tomamos el conjunto x_2, y_2, y generamos un grafico de dispersion

ax2.scatter(x_2,y_2,color='#fcba03')

# Creamos un grafico de barras

ax3.bar(comidas.keys(),comidas.values(),color='#03c6fc')

# Por ultimo un histograma

ax4.hist(np.random.randn(1000),color='#fc036b')
```
